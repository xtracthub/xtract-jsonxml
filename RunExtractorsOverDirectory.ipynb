```python
import os, json
import xtract_jsonxml_main 
from json.decoder import JSONDecodeError
from xml.parsers.expat import ExpatError
```

```python
dir_path = "../CDIACPub8"
```

```python
index = 0
for subdir, dirs, files in os.walk(dir_path):
    for file in files:
        file_path = os.path.join(subdir, file)
        metadata = dict()
        try:
            metadata = xtract_jsonxml_main.extract_json_metadata(file_path)
        except UnicodeDecodeError:
            print(file_path, "cannot be decoded")
            pass
        except JSONDecodeError:
            print(file_path, "not a valid json/xml")
            pass
        except ExpatError:
            print(file_path, "not well formed")
            pass
        except TypeError:
            print(file_path, "type error")
            pass
        
        json_path = "CDIACJSONXMLExtracted/{fn}JSONXMLXtract{idx}.json".format(fn=file, idx=index)
        with open(json_path, "w+") as outfile:
            json.dump(metadata, outfile)
        index += 1

print(index)
```
